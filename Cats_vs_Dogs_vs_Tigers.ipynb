```python
train_dir= "/Users/elena/Documents/CNN/dataset/training_set"
test_dir = "/Users/elena/Documents/CNN/dataset/test_set"

print(os.listdir(train_dir + "/cats"))
```

```python
import cv2

image = cv2.imread(train_dir+"/cats"+"/cat.1.jpg")

import matplotlib.pyplot as plt
plt.imshow(image)  # This shows wrong color (BGR)



```

```python
#3. Data Preprocessing
# ပုံတွေရဲ့ Range  ကို scale လုပ်တာ။
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255.
datagen_train = ImageDataGenerator( rescale = 1.0/255. )
datagen_test  = ImageDataGenerator( rescale = 1.0/255. )
```

```python
#2. ImageDataGenerator မှာ flow_from_directory() မှာ shuffle=True for train, shuffle=False for test ✅
# Image Data Preprocessing Challenge
batch_size = 5
train_set = datagen_train.flow_from_directory(train_dir,
                                            target_size = (100,100),
                                            #color_mode="grayscale",
                                             batch_size = batch_size,
                                            class_mode = "categorical", #if more than 2 categories: categorical, 2 categories: binary
                                            shuffle = True)
test_set = datagen_test.flow_from_directory(test_dir,
                                       target_size = (100,100),
                                       #color_mode="grayscale",
                                       batch_size = batch_size,
                                       class_mode = "categorical",
                                       shuffle = False)
```

```python
#CNN Model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(1000, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python
from tensorflow.keras.utils import load_img

img = load_img('/Users/elena/Documents/CNN/dataset/training_set/cats/cat.1.jpg')
print(img)
```

```python
model.compile(optimizer='adam',   #result ပိုေကာင်းအောင်ထည့်တာ
              loss = 'categorical_crossentropy',
              metrics =['acc'])   #accuracy
```

```python
#Training
# Data Training
#num_of_training_images mharကိုယ့်ရဲ့ training data ပုံအရေအတွက် ပြောင်းဖို့မမေ့ပါနဲ့
# CNN Model Training Process
# Understanding Training Parameters and Batch Processing

# num_of_training_images = 15
# batch_size=3
# history = model.fit(train_set,
#                     steps_per_epoch=num_of_training_images//batch_size, #5steps
#                     epochs=100,                   #count of runnning
#                     verbose = 2,
#                     shuffle = False)  #data တွေကို random/shuffle လုပ်ပြီး run

```

```python
batch_size = 32  # Larger batch sizes = faster training, but needs more memory

steps_per_epoch = train_set.samples // batch_size
validation_steps = test_set.samples // batch_size

history = model.fit(
    train_set,
    steps_per_epoch=steps_per_epoch,
    epochs=10,
    validation_data=test_set,
    validation_steps=validation_steps,
    verbose=2,
    shuffle=True
)
```

```python
#Evaluation
# Class Labels
class_dictionary = test_set.class_indices
print('Labels dictionary',class_dictionary)  #evaluation စမ်းရင် unseen testing data နဲ့စမ်း

y_predicted = model.predict(test_set) # A way to predict data

y_pred = []
for i in  y_predicted:
  print(i)
  print(np.argmax(i))
  print()
  y_pred.append(np.argmax(i))
```

```python
y_test = test_set.classes.tolist()
y_test
```

```python
# Predictions and Evaluation
#y_predicted = model.predict(test_set)
y_pred = [np.argmax(i) for i in y_predicted]
y_test = test_set.classes.tolist()
```

```python
# Confusion Matrix + Metrics
# from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

print('Accuracy:', accuracy_score(y_test, y_pred) * 100, '%')
print('Precision (macro):', precision_score(y_test, y_pred, average='macro') * 100, '%')
print('Recall (macro):', recall_score(y_test, y_pred, average='macro') * 100, '%')
print('F1 Score:', f1_score(y_test, y_pred, average='macro') * 100, '%')

```
